# Testing code for SLURM submissions

```python
import yaml
import itertools
import os
```

```python
def load_config(yaml_path: str):
    """Load the YAML config file and return the parsed dictionary."""
    with open(yaml_path, 'r') as f:
        config = yaml.safe_load(f)
    return config
```

```python
config_path = "../config/config.yaml"
config = load_config(config_path)
print(config)
```

```python
config['main']
```

```python
def generate_model_configs(config: dict):
    """Generate all possible combinations of model configurations."""
    if 'main' in config:
        config = config['main']
    categories = config.keys()  # Assuming your YAML has categories as top-level keys
    values = [config[category] for category in categories]  # Extract values for each category
    return list(itertools.product(*values))  # Generate combinations
```

```python
generate_model_configs(config)
```

```python
def create_slurm_script(build: str, analysis_date: str, country: str, model_type: str) -> str:
    """Create a Slurm batch script for this job configuration."""
    # Define paths
    parent_dir = os.path.abspath(os.path.join(os.getcwd(), ".."))
    data_dir = parent_dir + f"/data/{build}/time-stamped/{analysis_date}/"
    output_dir = parent_dir + f"/results/{build}/estimates/{model_type}/"

    # Make sure the output directory exists
    os.makedirs(output_dir, exist_ok=True)
    # Print the paths
    print(f"Data directory: {data_dir}")
    print(f"Output directory: {output_dir}")
    # Slurm script template
    slurm_script = f"""#!/bin/bash
#SBATCH --job-name=fit_model_{model_type}_{country}_{analysis_date}
#SBATCH --output={output_dir}/job_{country}_{analysis_date}.out
#SBATCH --error={output_dir}/job_{country}_{analysis_date}.err
#SBATCH --time=02:00:00
#SBATCH --ntasks=1
#SBATCH --cpus-per-task=4
#SBATCH --mem=16GB

# Activate virtual environment (if needed)
# mamba activate antigen

# Run the model fitting script
python fit_model.py --data_path {data_dir} --country {country} --model {model_type} --output_dir {output_dir}
    """

    # Save Slurm script
    slurm_script_path = os.path.join(output_dir, f"job_{country}_{analysis_date}.sh")
    with open(slurm_script_path, 'w') as f:
        f.write(slurm_script)
    print(f"Slurm script saved to {slurm_script_path}")
    return slurm_script_path
```

```python
# Test the function
create_slurm_script("flu-simulated-150k-samples", "2021-01-01", "south", "MLR")
```

```python
def main():
    # Paths
    seq_counts_path = 'seq_counts.tsv'
    case_counts_path = 'case_counts.tsv'
    config_path = 'config.yaml'
    output_base_dir = 'output_results'

    # Load the YAML configuration
    config = load_config(config_path)

    # Generate all combinations of model configurations
    model_configs = generate_model_configs(config)

    # Submit each configuration as a Slurm job
    for job_id, model_config in enumerate(model_configs):
        model_output_dir = os.path.join(output_base_dir, f"config_{job_id}")
        os.makedirs(model_output_dir, exist_ok=True)

        # Create Slurm batch script for this model configuration
        slurm_script_path = create_slurm_script(seq_counts_path, case_counts_path, model_config, model_output_dir, job_id)

        # Submit the job to the cluster
        os.system(f"sbatch {slurm_script_path}")  # Slurm job submission
```
